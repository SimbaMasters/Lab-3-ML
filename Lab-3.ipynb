```python
# make a table explaining the columsn and their values

feature_description = pd.DataFrame()

feature_description['Features'] = df_cleaned.columns

# TODO: the following commented out sections are columns that need to be 
# populated for the feature description table

#feature_description['Description'] = ["""populate this"""]

#feature_description['Scales'] = ["""populate this"""]

#feature_description['Discrete/Continuous'] = ["""populate this"""]


# function that simplifies the process of finding the value ranges of the features
def rangeList(df: pd.DataFrame) -> list[any]:
    arr = list()
    for col in df.columns:
        if type(df[col][0]) == str:
            arr.append(df[col].unique())
        else:
            arr.append(f"{df[col].min()} - {df[col].max()}")
    return arr

feature_description['Ranges'] = rangeList(df_cleaned)

pd.set_option('display.max_colwidth', None)


"""uncomment the line below if you want to see the table"""
#feature_description
```
